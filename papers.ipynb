## Rubric for grading

I'll grade you based on the following criteria:

- Did you state the problem?
- Did you motivate the problem -- why was the problem important to solve?
- Did you explain why the problem was hard?
- Did you briefly summarize previous progress on the problem?
- Did you briefly summarize what you will talk about?
- Was the presentation easy to follow?
    - Was the use of notation consistent?
    - Were quantities/concepts adequately defined?
    - Were there enough logical steps provided in derivations / was there intuition built for the answer (if full derivation was too long)?
    - Was the connection between parts of the presentation clear?
    - Did you interpret the results?
- Was the presentation stimulating?
    - Did you show a code demo?
    - Did you show figure(s)?
- Did you discuss the limitations of the method/solution presented, if any?
- Did you give a brief future outlook?
