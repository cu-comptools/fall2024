# Presentations

## Logistics

1. Pick 3 papers from the selection, or suggest your own (see below).
2. Email me with your selection, in descending order of preference, by **Tuesday October 1st**.
3. You'll get your paper allocated by Wednesday, October 2nd.
4. You'll also get a randomly drawn day between October 11-21 for your presentation. The order of presentations will be reversed for the second round (towards the end of the semester) for fairness.
5. When you're not presenting, you'll still have to read the paper and will get ~3 questions to answer about the paper being presented. There will be no other homework given during this time.
6. I'll start with a presentation on October 9th. You should use the structure of my presentation to guide you in your own.

## Rubric for grading

You may use slides or the blackboard for your presentation.
Your grade will depend on the following criteria:

- Did you state the problem?
- Did you motivate the problem -- why was the problem important to solve?
- Did you explain why the problem was hard?
- Did you briefly summarize previous progress on the problem?
- Did you briefly summarize what you will talk about?
- Was the presentation easy to follow?
    - Was the use of notation consistent?
    - Were quantities/concepts adequately defined?
    - Were there enough logical steps provided in derivations / was there intuition built for the answer (if full derivation was too long)?
    - Was the connection between parts of the presentation clear?
    - Did you interpret the results?
- Was the presentation stimulating?
    - Did you show a code demo?
    - Did you show figure(s)?
- Did you discuss the limitations of the method/solution presented, if any?
- Did you give a brief future outlook?
